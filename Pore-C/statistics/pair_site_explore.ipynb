```python
import pandas as pd
from pathlib import Path
from pathlib import PurePath
```

```python
#the same code as pair readin as df in porecplot_pre_func.py v2.2
def read_pair(porec_dir,prefix):
    from pathlib import Path
    from pathlib import PurePath
    pair_dir = Path(PurePath(porec_dir,'pairs'))
    pair_file = str(list(pair_dir.glob(prefix + "*.sorted.pairs.gz"))[0])
    pair_table = pd.read_table(
        pair_file,                     
        comment="#",             
        compression='gzip',
        names=['readID','chr1','pos1','chr2','pos2','strand1','strand2','pair_type','align1_idx','align2_idx','distance_on_read']
    )

    pair_table.replace({'readID': {'read|_[0-9]+': ''}},regex=True,inplace=True)
    return pair_table
```

```python
def pair_anchor(anchor_df,pair_df):
    pair_pos1_anchor_df = pd.DataFrame([])
    pair_pos2_anchor_df = pd.DataFrame([])
    for i in anchor_df.itertuples():
        anchor_index,anchor_chr,anchor_start,anchor_end,anchor_name = i

        df_pair_pos1_anchor = pair_df.query(
            'chr1 == @anchor_chr and pos1 <= @anchor_end and pos1 >= @anchor_start'
        ).loc[:,('readID', 'chr1', 'pos1', 'strand1','pair_type', 'align1_idx','distance_on_read')]

        df_pair_pos1_anchor.loc[:,'anchor1'] = anchor_name
        pair_pos1_anchor_df = pair_pos1_anchor_df.append(df_pair_pos1_anchor)

        df_pair_pos2_anchor = pair_df.query(
            'chr2 == @anchor_chr and pos2 <= @anchor_end and pos2 >= @anchor_start'
        ).loc[:,('readID', 'chr2', 'pos2', 'strand2','pair_type', 'align2_idx','distance_on_read')]

        df_pair_pos2_anchor.loc[:,'anchor2'] = anchor_name
        pair_pos2_anchor_df = pair_pos2_anchor_df.append(df_pair_pos2_anchor)
    pair_anchor_df = pd.merge(pair_pos1_anchor_df,pair_pos2_anchor_df,how='outer')
    pair_anchor_df_nona = pair_anchor_df.dropna()
    return pair_anchor_df_nona
```

# 使用anchor_detial 获取对site去重后的pair

```python
%%writefile '/public/home/lizw/task/pore_c/tools/statistics/anchor_link.py'
#v1 --anchor_file base on the anchor_detail generated by porecplot_func v2.4
import click 
import itertools
import pandas as pd
@click.command()
@click.option('--anchor_file',type=str)
@click.option('--comb_file',type=str)
@click.option('--color',type=str)
@click.option('--prefix',type=str)
@click.option('--path',type=str)
@click.option('--sample',type=float)
def anchor_dropdu(anchor_file,comb_file,color,prefix,path,sample):
    anchor_reads_detial_df = pd.read_table(anchor_file,index_col=0)
    df_comb = pd.read_csv(comb_file)
    anchor_and2_list = df_comb['read_name'].to_list()
    anchor_and2_detail = anchor_reads_detial_df.query('read_name in @anchor_and2_list')
    #only calculte the combination, would not calculate if the interaction caused by repeats of same anchor sites
    anchor_and2_detail_dropdu = anchor_and2_detail.drop_duplicates(['read_name','anchor_name'])
    detial_dropdu_combinations = anchor_and2_detail_dropdu.groupby('read_name').apply(lambda x : list(itertools.combinations(x.index,2)))
    anchor_comb_list = list(itertools.chain(*detial_dropdu_combinations))
    df_anchor = pd.DataFrame([])
    for i in anchor_comb_list:
        df1 = pd.DataFrame(anchor_and2_detail_dropdu.loc[i[0],:]).T.reset_index()
        df2 = pd.DataFrame(anchor_and2_detail_dropdu.loc[i[1],:]).T.reset_index()
        df_temp = pd.concat([df1,df2],axis=1)
        df_anchor = df_anchor.append(df_temp)
    df_link = df_anchor.iloc[:,[1,3,3,6,8,8]]
    df_link.iloc[:,1] = df_link.iloc[:,1] -1
    df_link.iloc[:,4] = df_link.iloc[:,4] -1
    df_link.iloc[:,0] = ['tair' + str(a) for a in df_link.iloc[:,0]]
    df_link.iloc[:,3] = ['tair' + str(a) for a in df_link.iloc[:,3]]
    df_link['color'] = f'color={color}'
    df_anchor.sample(frac=sample).to_csv(f'{path}/{prefix}_anchor_pair.tsv',sep='\t')
    df_link.sample(frac=sample).to_csv(f'{path}/{prefix}_link.txt',sep=' ',header=None,index=None)
if __name__ == '__main__':
    anchor_dropdu()
```

```python
%%writefile '/public/home/lizw/task/pore_c/tools/statistics/anchor_link.py'
#@Author: Zhuowen Li
#@LastEdit: 2021/8/28 下午4:58:05
#@Version: v2 --anchor_file base on the anchor filter  generated by porecplot_func v2.5
#@Description: 
import click 
import itertools
import pandas as pd
@click.command()
@click.option('--anchor_file',type=str)
@click.option('--comb_file',type=str)
@click.option('--color',type=str)
@click.option('--prefix',type=str)
@click.option('--path',type=str)
@click.option('--sample',type=float)
def anchor_dropdu(anchor_file,comb_file,color,prefix,path,sample):
    anchor_reads_detial_df = pd.read_table(anchor_file,index_col=0)
    df_comb = pd.read_csv(comb_file)
    anchor_and2_list = df_comb['read_name'].to_list()
    anchor_and2_detail = anchor_reads_detial_df.query('read_name in @anchor_and2_list')
    
    #only calculte the combination, would not calculate if the interaction caused by repeats of same anchor sites
    anchor_and2_detail_dropdu = anchor_and2_detail.drop_duplicates(['read_name','anchor_name'])
    detial_dropdu_combinations = anchor_and2_detail_dropdu.groupby('read_name').apply(lambda x : list(itertools.combinations(x.index,2)))
    anchor_comb_list = list(itertools.chain(*detial_dropdu_combinations))
    
    df_anchor = pd.DataFrame([])
    for i in anchor_comb_list:
        df1 = pd.DataFrame(anchor_and2_detail_dropdu.loc[i[0],:]).T.reset_index()
        df2 = pd.DataFrame(anchor_and2_detail_dropdu.loc[i[1],:]).T.reset_index()
        df2.columns=['index2', 'chrom2', 'read_name2', 'fragment_mid2', 'anchor_name2']
        df_temp = pd.concat([df1,df2],axis=1)
        df_anchor = df_anchor.append(df_temp)

    df_anchor_anchorpair = df_anchor.query('(anchor_name != "normal") and (anchor_name2 != "normal")')
    df_anchor_normal = df_anchor.query('(anchor_name == "normal") or (anchor_name2 == "normal")')
    df_anchor_anchorpair['color'] = f'color={color}'
    df_anchor_normal['color'] = 'color=grey'
    
    df_anchor_color = pd.concat([df_anchor_anchorpair,df_anchor_normal],axis=0)
    
    df_link = df_anchor_color.iloc[:,[1,3,3,6,8,8,10]]
    df_link.iloc[:,1] = df_link.iloc[:,1] -1
    df_link.iloc[:,4] = df_link.iloc[:,4] -1
    df_link.iloc[:,0] = ['tair' + str(a) for a in df_link.iloc[:,0]]
    df_link.iloc[:,3] = ['tair' + str(a) for a in df_link.iloc[:,3]]
    
    df_anchor.sample(frac=sample).to_csv(f'{path}/{prefix}_anchor_pair.tsv',sep='\t')
    df_link.sample(frac=sample).to_csv(f'{path}/{prefix}_link.txt',sep=' ',header=None,index=None)
if __name__ == '__main__':
    anchor_dropdu()
```

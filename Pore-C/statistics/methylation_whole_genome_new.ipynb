```python
%config InlineBackend.figure_format = 'retina'

from matplotlib import pyplot as plt
import matplotlib.ticker as ticker
import matplotlib.font_manager as font_manager
import matplotlib as mpl
import matplotlib.patches as mp
import seaborn as sns
```

```python
import pandas as pd
import numpy as np
```

```python
Long_C_CG_500k = pd.read_csv('/public/home/lizw/task/pore_c/methylation/deepsignal_plant/merge_CG_500K_500000_level_size_in_interval.csv',converters={'chrom':str})
Long_C_CG_500k['start']  = [int(x[0][1:]) for x in Long_C_CG_500k['interval'].str.split(',')]
Long_C_CG_500k['end'] = [int(x[1][:-1]) for x in Long_C_CG_500k['interval'].str.split(',')]

Long_C_CHG_500k = pd.read_csv('/public/home/lizw/task/pore_c/methylation/deepsignal_plant/merge_CHG_500K_500000_level_size_in_interval.csv',converters={'chrom':str})
Long_C_CHG_500k['start']  = [int(x[0][1:]) for x in Long_C_CHG_500k['interval'].str.split(',')]
Long_C_CHG_500k['end'] = [int(x[1][:-1]) for x in Long_C_CHG_500k['interval'].str.split(',')]

Long_C_CHH_500k = pd.read_csv('/public/home/lizw/task/pore_c/methylation/deepsignal_plant/merge_CHH_500K_500000_level_size_in_interval.csv',converters={'chrom':str})
Long_C_CHH_500k['start']  = [int(x[0][1:]) for x in Long_C_CHH_500k['interval'].str.split(',')]
Long_C_CHH_500k['end'] = [int(x[1][:-1]) for x in Long_C_CHH_500k['interval'].str.split(',')]
```

```python
chrom_size = {
    '1': [30_000_000, 30],
    '2': [20_000_000, 20],
    '3': [25_000_000, 25],
    '4': [20_000_000, 20],
    '5': [30_000_000, 30],
}

font_dirs = ['/public/home/mowp/test/fonts/']
font_files = font_manager.findSystemFonts(fontpaths=font_dirs)

for font_file in font_files:
    font_manager.fontManager.addfont(font_file)

plt.rcParams['font.family'] = 'Arial'
plt.rcParams['font.size'] = 30
plt.rcParams['svg.fonttype'] = 'none'
fig, ax = plt.subplots(nrows=1, ncols=6, figsize=(30,8), sharey=True)
for i in ['1', '2', '3', '4', '5']:
    ax_chr = ax[int(i) - 1]
    df = Long_C_CG_500k.query('chrom == @i')
    ax_chr.plot(df['start'], df['met_level'], color='#262180',lw=2)
    ax_chr.set_xticks([0, chrom_size[i][0]])
    ax_chr.set_xticklabels([0, chrom_size[i][1]])
    ax_chr.title.set_text(f'Chr{i}')
    sns.despine(ax=ax_chr)

for i in ['1', '2', '3', '4', '5']:
    ax_chr = ax[int(i) - 1]
    df = Long_C_CHG_500k.query('chrom == @i')
    ax_chr.plot(df['start'], df['met_level'], color='#4C7C5D',lw=2)
    ax_chr.set_xticks([0, chrom_size[i][0]])
    ax_chr.set_xticklabels([0, chrom_size[i][1]])
    ax_chr.title.set_text(f'Chr{i}')
    sns.despine(ax=ax_chr)
    
for i in ['1', '2', '3', '4', '5']:
    ax_chr = ax[int(i) - 1]
    df = Long_C_CHH_500k.query('chrom == @i')
    ax_chr.plot(df['start'], df['met_level'], color='#632523',lw=2)
    ax_chr.set_xticks([0, chrom_size[i][0]])
    ax_chr.set_xticklabels([0, chrom_size[i][1]])
    ax_chr.title.set_text(f'Chr{i}')
    sns.despine(ax=ax_chr)

from matplotlib.lines import Line2D
custom_lines = [Line2D([0], [0], color='#262180', lw=2),
                Line2D([0], [0], color='#4C7C5D', lw=2),
                Line2D([0], [0], color='#632523', lw=2)]


ax[5].legend(custom_lines, ['CG','CHG','CHH'],loc='upper right')
ax[5].spines['right'].set_color('none')
ax[5].spines['bottom'].set_color('none')
ax[5].spines['left'].set_color('none')
ax[5].spines['top'].set_color('none')
ax[5].set_xticks([])
ax[5].set_yticks([])


plt.ylim(0, 1)
plt.yticks([0, 1])
ax[0].set_ylabel('Methylation Level')

# chr5子图横轴，最后自己手动加个 (Mb吧)
plt.savefig('/public/home/lizw/task/pore_c/tools/from_wp/methylation_geno_500k.png',format='png',dpi=300,bbox_inches='tight')
```

```python
#获取对sub peak贡献top20的比对对应的fragment的序列
import pandas as pd
reads_info = pd.read_table('/public/home/lizw/task/pore_c/porec_1000_filter_mainchr_result/porecplot/DpnII_run04.order_3--sub_peak_binsize500000_ylimno_and_binnor_bar_anchor_filter_merge.csv',converters={'chr':str})

reads_info

sub_peak = pd.read_table('/public/home/lizw/task/pore_c/porec_1000_filter_mainchr_result/porecplot/sub_peak_peri/sub_peak_interacting.csv',header=None,names=['interval','chr'],converters={'chr':str})

sub_peak_reads = pd.merge(reads_info,sub_peak,on=['chr','interval'],how='right')

sub_peak_reads.groupby(['chr','interval','fragment_mid']).count()

sub_peak_reads['count'] = sub_peak_reads.groupby(['chr','interval','fragment_mid'])['read_name'].transform('count')

sub_peak_reads.groupby(['chr','interval','fragment_mid'])['read_name'].pipe(lambda g:(g.count()/g.count().sum())).sort_values(ascending=False)

####要看看这些贡献最多的位置是不是重复序列看看比对上这些区域的reads比对是什么样子的

frag = pd.read_parquet('/public/home/lizw/task/pore_c/porec_1000_filter_mainchr_result/virtual_digest/DpnII_TAIR10.vd.fragments.parquet')

frag['fragment_mid'] = frag.eval('end+start+1')//2

sub_peak_reads_frag_count = pd.merge(sub_peak_reads,frag,how='left',left_on=['chr','fragment_mid'],right_on=['chrom','fragment_mid']) #add in frag genome location inf on the sub_peak reads 

top_20 = sub_peak_reads_frag_count.reindex(['chr','start','end','count','interval','fragment_mid','fragment_id'],axis=1).sort_values(by='count',ascending=False).drop_duplicates().head(20)

import pysam

def get_seq(row,fa_file):
    fasta_open = pysam.Fastafile(fa_file)
    sub_seq = fasta_open.fetch(row['chr'],row['start'],row['end'])
    fasta_open.close()
    return sub_seq

fa_path = '/public/home/lizw/data/IMPORTANT_genome/arabidopsis/tair10/tair10.simplify.fa'
top_20['seq'] = top_20.apply(get_seq,axis=1,fa_file = fa_path)

from Bio import SeqIO
from Bio.Seq import Seq
from Bio.SeqRecord import SeqRecord

record = []
for j in top_20.itertuples():
    seq_id =str(j.fragment_id)
    seq = j.seq
    record.append(SeqRecord(Seq(seq),id=seq_id))

SeqIO.write(record,'/public/home/lizw/task/pore_c/porec_1000_filter_mainchr_result/porecplot/sub_peak_peri/top20.fa','fasta')
```

```python
align_sub = align_dir_porec_generator_df.loc[:,('read_idx', 'align_idx', 'align_type', 'chrom', 'start', 'end','strand', 'read_name', 'read_length', 'read_start', 'read_end','mapping_quality','filter_reason','fragment_id')]
```

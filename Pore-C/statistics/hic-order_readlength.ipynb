```python
#merge
import matplotlib.pyplot as plt
import matplotlib.font_manager as font_manager
import matplotlib.ticker as ticker
import seaborn as sns
import pandas as pd
font_dirs = ["/public/home/lizw/software/font"]
font_files = font_manager.findSystemFonts(fontpaths=font_dirs)
for font_file in font_files:
    font_manager.fontManager.addfont(font_file)
plt.rcParams["font.family"] = "Arial"
plt.rcParams["font.size"] = 50
f, ax = plt.subplots(1,figsize=(12,16))
df_length = pd.read_table('/public/home/lizw/task/pore_c/fastq/merge_porec_new/nanoplot/merge_porec_new_seabornNanoPlot-data.tsv.gz',compression='gzip')
binsize=1000
df_cut,df_bin = pd.cut(df_length['lengths'],bins=range(0,df_length['lengths'].max()+binsize-1,binsize),precision=0,retbins=True)
df_count = df_cut.value_counts()
df_count.head(10).plot(kind='bar',ax=ax,color='#3C8C6F',width=0.9,ec='white')
sns.despine(ax=ax)
plt.xticks(ticks=np.arange(0,10,2)+0.5,labels=[f'{x}k' for x in np.arange(1000,11000,2000)//1000],rotation=30)
plt.yticks(ticks=np.arange(0,3000000+500000,500000),labels=[f'{x}M' for x in np.arange(0,3000000+500000,500000)/1000000])
ax.set_xlabel('Read Length')
ax.set_ylabel('Read Count')
plt.savefig('/public/home/lizw/task/pore_c/fastq/merge_porec_new/merge_porec_new_length_dis.png',format='png',dpi=300,bbox_inches='tight')
#ax.yaxis.set_major_formatter(ticker.FormatStrFormatter('%d'))
```
